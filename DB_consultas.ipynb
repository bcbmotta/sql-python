```python
import os
from sqlalchemy import create_engine
import pandas as pd
```

```python
def consulta(query):
    return pd.read_sql(query, conn)
```

```python
# Env variables
USERNAME = os.getenv('USERNAME')
PASSWORD = os.getenv('PASSWORD')
HOST = os.getenv('HOST')
PORT = os.getenv('PORT')
DATABASE = os.getenv('DATABASE')

# Define connection string and create engine
string_conn = f'postgresql://{USERNAME}:{PASSWORD}@{HOST}:{PORT}/{DATABASE}'
conn = create_engine(string_conn)
```

```python
SQL = '''SELECT table_name, column_name, data_type
         FROM information_schema.columns 
         WHERE table_schema = 'public' 
         ORDER BY table_name, ordinal_position;'''
consulta(SQL)
```

1. Faça uma consulta que retorne todos os vendedores do estado do Paraná.

```python
SQL = '''
         SELECT *
         FROM sellers
         WHERE seller_state = 'PR';
      '''
consulta(SQL)
```

2. Faça uma consulta que retorne todos os vendedores do estado do Paraná ou da cidade de sarandi.

```python
SQL = '''
         SELECT *
         FROM sellers
         WHERE seller_state = 'PR'
            OR seller_city = 'sarandi';
      '''
consulta(SQL)
```

3. Faça uma consulta que retorne todos os pedidos que o preço esteja entre R\$ 45 e R\$ 125.

```python
SQL = '''
         SELECT *
         FROM order_payments
         WHERE payment_value BETWEEN 45 AND 125;
      '''
consulta(SQL)
```

4. Faça uma consulta que retorne todos os pedidos calculando o valor total do pedido (preço + frete).

```python
SQL = '''
         SELECT o.order_id, oi.price, oi.freight_value, (price + freight_value) AS total
         FROM order_items oi
         JOIN orders o ON oi.order_id = o.order_id;
      '''
consulta(SQL)
```
